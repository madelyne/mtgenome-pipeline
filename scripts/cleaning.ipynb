```python
from Bio import SeqIO
from Bio.Align.Applications import MuscleCommandline
from Bio import Align
import numpy as np
import math
##### 
#
#
#
#



# >>> muscle_exe = r"C:\Program Files\Aligments\muscle3.8.31_i86win32.exe"
# >>> in_file = r"C:\My Documents\unaligned.fasta"
# >>> out_file = r"C:\My Documents\aligned.fasta"
# >>> muscle_cline = MuscleCommandline(muscle_exe, input=in_file, out=out_file)
# >>> print(muscle_cline)

# fetch working directory from user
file_path = input('Enter file path to current work directory: ')


seq_read = SeqIO.read(file_path + 'data/data1.txt', 'fasta')
ref_read = SeqIO.read(file_path + 'data/ref.txt', 'fasta')

aligner = Align.PairwiseAligner()

aligner.mode = 'local'

seq1 = seq_read.seq
seq2 = ref_read.seq

score = aligner.score(seq1, seq2)
print(score)

alignments = aligner.align(seq1, seq2)
for alignment in alignments:
    print(alignment)


# fetch reference file name from user
ref_name = input('Enter reference file name: ')


## the below code calculates the k2p distance matrix from FASTA files
seqs = []
my_array = []
for i in range(1, 10):
	for record in SeqIO.parse('data' + str(i)+ '.txt', 'fasta'):
		seqs.append(record.seq)


#calculate proportion of transitions
def p_score(seq1, seq2, cap):
	p_count = 0
	for i in range(0, cap):
		if (seq1[i] == 'a' and seq2[i] == 'g') \
            or (seq1[i] == 'g' and seq2[i] == 'a') \
            or (seq1[i] == 't' and seq2[i] == 'c') \
            or (seq1[i] == 'c' and seq2[i] == 't'):
				p_count += 1
				print(p_count, cap, p_count / cap)
				return (p_count / cap)


#calculate proportion of transversions
def q_score(seq1, seq2, cap):
	q_count = 0
	for j in range(0, cap):
		if (seq1[j] == 'a' and seq2[j] == 'c') or (seq1[j] == 'c' and seq2[j] == 'a') or (seq1[j] == 'c' and seq2[j] == 'g') or (seq1[j] == 'g' and seq2[j] == 'c') or (seq1[j] == 'a' and seq2[j] == 't') or (seq1[j] == 't' and seq2[j] == 'a') or (seq1[j] == 'g' and seq2[j] == 't') or (seq1[j] == 't' and seq2[j] == 'g'):
			q_count += 1
	print(q_count, cap, q_count / cap)
	return (q_count / cap)


#calculate final k2p score
def k2p(seq1, seq2):
	if len(seq1) < len(seq2): #cap length of comparison to shorter of two seqs
		cap = len(seq1)
	else:
		cap = len(seq2)
	p = p_score(seq1, seq2, cap)
	q = q_score(seq1, seq2, cap)
	score = -0.5 * np.log((1 - 2*p - q) * math.sqrt(1 - 2*q))
	#print(p, q, score)
	return (-0.5 * np.log((1 - 2*p - q) * math.sqrt(1 - 2*q)))


#slot k2p values into distance matrix array
def dist_mat(seqs):
	my_array = np.empty([9, 9])
	for i in range(len(seqs)):
		for j in range(len(seqs)):
			my_array[i,j] = k2p(seqs[i].lower(), seqs[j].lower())
			my_array[i,j] = round(my_array[i,j],4)
			np.savetxt('tablefile.txt', my_array, fmt='%.4f')
	print(my_array)
	return my_array

dist_mat(seqs)
#
# /Users/madelyne/Desktop/mtgenome-pipeline/
```
